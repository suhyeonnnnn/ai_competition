```python
# 필요없는 변수 drop
train = train.drop(['index'], axis=1)
test = test.drop(['index'], axis=1)
```

```python
def age_grouping(x):
    if x <= 10:
        return 10
    elif x <= 20:
        return 20
    elif x <= 30:
        return 30
    elif x <= 40:
        return 40
    elif x <= 50:
        return 50
    elif x <= 60:
        return 60
    elif x <= 70:
        return 70
    elif x <= 80:
        return 80
    elif x <= 90:
        return 90
    elif x <= 100:
        return 100
    else: 
        return x
```

```python
train.age = train.age.apply(age_grouping)
test.age = test.age.apply(age_grouping)
```

```python
train.age.value_counts()
```

```python
test.age.value_counts()
```

```python
# 100 이상값 삭제
idx1 = train[train['age'] > 100].index
idx2 = test[test['age'] > 100].index

train = train.drop(idx1)
test = test.drop(idx2)
```

```python
# country
# 100이하면 '기타'로처리
country_list = train.country.value_counts()[:20]
def func(country):
    if country in country_list:
        return country
    else:
        return '기타'

train['country'] = train['country'].apply(lambda x : func(x))
test['country'] = test['country'].apply(lambda x : func(x))
```
