# 나이, 가족수 이상치 처리

#### 나이 답변을 이상하게 한 사람은은 nerd일까?

크게 상관없는듯 -> 결측치, 이상치(2999?)는 평균으로 채우기로 결정

```python
def age_outlier(x):
    if x > 100:
        return round(train.age.mean())
    else: 
        return x
```

```python
train.age = train.age.apply(age_outlier)
test.age = test.age.apply(age_outlier)
```

```python
def age_grouping(x):
    if x <= 10:
        return 10
    elif x <= 20:
        return 20
    elif x <= 30:
        return 30
    elif x <= 40:
        return 40
    elif x <= 50:
        return 50
    elif x <= 60:
        return 60
    elif x <= 70:
        return 70
    elif x <= 80:
        return 80
    elif x <= 90:
        return 90
    elif x <= 100:
        return 100
    else: 
        return x
```

```python
train.age = train.age.apply(age_grouping)
test.age = test.age.apply(age_grouping)
```
